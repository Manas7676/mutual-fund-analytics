## Insight 1 – Historical Risk (VaR & CVaR)

Small-cap and high-risk equity funds exhibit the largest downside risk based on Historical VaR (95%) and CVaR. These funds experience larger losses during extreme market movements compared to large-cap or debt-oriented funds.

## Insight 2 – Rolling Sharpe Ratio

The rolling 90-day Sharpe Ratio highlights periods of superior risk-adjusted performance. Large-cap funds generally maintain more stable Sharpe Ratios, while small-cap funds show higher volatility.

## Insight 3 – Investor Cohort Analysis

Recent investor cohorts contribute higher overall investment amounts, reflecting increased participation in mutual funds over recent years. Cohort analysis also identifies the most preferred schemes among new investors.

## Insight 4 – SIP Continuity

Investors with average SIP gaps greater than 35 days are classified as At-Risk. This analysis can help mutual fund companies proactively identify investors who may discontinue systematic investments.

## Insight 5 – Portfolio Diversification

Funds with higher Herfindahl–Hirschman Index (HHI) values maintain more concentrated portfolios, while lower HHI values indicate better diversification across holdings and sectors.
